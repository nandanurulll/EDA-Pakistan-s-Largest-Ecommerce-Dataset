### Cek Outliers

```python
# Def cek outliers
def Cek_Outliers(df, col):
    q1 = df[col].describe()['25%']
    q3 = df[col].describe()['75%']
    iqr = q3 - q1
    upper_fence = q3 + (1.5*iqr)
    lower_fence = q1 - (1.5*iqr)
    df_outliers = df[(df[col] < lower_fence) | (df[col] > upper_fence)]
    if len(df_outliers) == 0:
        return "No Outliers"
    else:
        return df_outliers
```

#### Kolom "Price"

```python
df_outliers_rat = Cek_Outliers(df,'price')
a = df_outliers_rat.count()['price']

iqr = df['price'].describe()['75%'] - df['price'].describe()['25%']
upper_fence = df['price'].describe()['75%'] + (1.5*iqr)
lower_fence = df['price'].describe()['25%'] - (1.5*iqr)

print("Terdapat", a, "data outlier pada kolom 'price'. Yang termasuk outlier adalah data yang lebih kecil dari", lower_fence, "atau yang lebih besar dari", upper_fence)
```

```python
# Visualisasi
plt.figure(figsize=(7,7))

# Boxplot untuk mengecek global outlier
plt.subplot(1,2,2)
sns.boxplot(df['price'])
plt.title("Boxplot kolom 'Price'")

plt.show()
```

Dari gambar diatas terlihat bahwa:
* Data 'price' yang termasuk dalam outlier adalah harga yang nilainya negatif dan positif tinggi. Untuk harga yang nilainya tinggi masih terbilang wajar atau masuk akal karena harga memang bisa saja nilainya tinggi sehingga tidak diperlukan penanganan. Namun, untuk data "price" yang bernilai negatif harus ditangani dengan cara memutlakkan (absolut) angka tersebut karena terdapat **kesalahan saat input data**.

**Penanganan kolom "price":**

```python
df['price'] = df['price'].apply(lambda x: abs(x))
```

#### 8. Kategori mana yang menghasilkan Penghasilan terbesar diantara Transaksi yang memiliki diskon

```python
# Bikin dataframe baru yang berisi daftar penjualan yang memiliki discount

a = df
a = pd.DataFrame(a)
a = a[a['discount_amount'] != 0.0]
```

```python
# membuat dataframe dari grouping dataframe a berdasarkan category, diambil nilai maksimum dari 'grand_total' dan diurutkan dari yang terbesar
X = a.groupby('category')['grand_total'].max().sort_values(ascending=False)
X = pd.DataFrame(X.reset_index())

# menggabungkan dataframe a dan X dengan kolom penghubung 'grand_total' dan 'category'
Z = pd.merge(X, a, on=['grand_total','category'])
Z = Z[['category','grand_total','discount_amount']]
Z = Z.drop_duplicates()
Z
```

```python
plt.figure(figsize=(15,7))  
ax = sns.barplot(data=Z, x='category', y='grand_total', color='Pink', label='Grand_total')
bx = sns.barplot(data=Z, x='category', y='discount_amount', color='Blue', label='Discount')
ax.bar_label(ax.containers[0])
bx.bar_label(bx.containers[1])
plt.legend()
plt.xticks(np.arange(len(X['category'].unique())), rotation=90)
plt.xlabel('Kategori' )
plt.ylabel('Nominal (Rupee Pakistan)')
plt.title('Penghasilan yang Memiliki Diskon Pada Tiap Kategori', size=16)
plt.savefig('Penghasilan_denga_diskon.png')
plt.show()
```

**Analisa:**<br>
Berdasarkan gambar grafik diatas terlihat bahwa kategori Mobiles & Tablets menghasilkan penjualan terbesar meskipun diskon atau potongan harga yang diberikan kecil, hanya 0,6%, sedangkan kategori Beauty & Grooming menempati urutan ke-8 penjualan terbesar padahal potongan harga yang diberikan cukup besar (sebesar 9,3%) dibandingkan kategori yang lain.

**Rekomendasi:**<br>
* Pada kategori yang paling banyak penjualan menandakan barang-tersebut banyak diminati, sebaiknya perbanyak/menambahkan jenis barang dan stock pada kategori tersebut menjadi lebih lengkap lagi, sehingga dapat menarik konsumen baru untuk membeli di eCommerce ini.
* Memberikan reward pada affiliator eCommerce yang telah berhasil mendatangkan pelanggan untuk berbelanja di eCommerce ini dengan tujuan mendatangkan affiliator-affiliator baru sehingga perusahaan dan affiliator sama-sama diuntungkan. Perusahaan mendapatkan keuntungan promosi/iklan sebagai salah satu strategi pemasaran gratis sehingga meminimalkan cost untuk promosi. Affiliator baru jika berhasil mengiklankan dan memikat konsumen hingga berbelanja di eCommerce ini akan mendapatkan komisi.
* Memberikan potongan harga atau potongan ongkos kirim untuk pelanggan yang menggunakan metode pembayaran yang perusahaan inginkan, misal metode pembayaran 'bankalfalah' dengan cara bekerja sama dengan bankalfalah, sehingga perusahaan dan bankalfalah sama-sama dapat diuntungkan. Perusahaan diuntungkan karena tidak perlu membuat tim khusus untuk mengelola *cash receipt* sehingga dapat menekan biaya operasional, dan bagi 'bankalfalah' mendapatkan nasabah baru.
* Untuk menarik konsumen baru untuk berbelanja di eCommerce ini adalah dengan membuat program event yang cocok di Pakistan seperti 'Black Friday' karena pada kenyataannya dengan adanya event tersebut sepertiga dari konsumen yg bergabung merupakan konsumen yang aktif berbelanja di eCommerce ini, dengan diadakan program/event baru diharapkan dapat mendatangkan lebih banyak konsumen baru dan meningkatkan penjualan perusahaan eCommerce.
* Kategori yang tidak begitu menghasilkan penjualan yang tinggi tidak perlu memberikan potongan harga yang besar, karena jika memberikan diskon yg cukup besar bisa jadi perusahaan tidak mendapatkan keuntungan. Pada kenyataannya kategori Mobiles & Tablets yang menjadi kategori menghasilkan penjualan tertinggi tidak memberikan diskonbesar, hanya 0.6%.
